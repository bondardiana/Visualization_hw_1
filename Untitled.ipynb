```python
import pandas as pd
import altair as alt

df = pd.read_csv('GLB.Ts_dSST.csv', sep=",")

da = df[["Year", "J-D"]]

#points
points = alt.Chart(da).mark_circle( size=80, 
    color='#f07330', 
    stroke='grey', 
    opacity = 1,
    strokeWidth=1,
    thickness=1).encode(
    x = alt.X(field = "Year", type = 'quantitative', title = "",  
            axis = alt.Axis(tickCount = 14, labelColor="Grey", labelFontSize=13),
            scale=alt.Scale(zero=False)),
    y = alt.Y(field ="J-D", type = "quantitative", title ="",
                            axis = alt.Axis(tickCount = 14, labelColor="Grey", labelFontSize=13)
))

#years_text
years_text = points.mark_text(align='left',baseline='middle',fontSize = 15, dx=-40
                                                ).encode(text = "Year:N").transform_filter(
    alt.FieldOneOfPredicate(field='Year', oneOf = [1904, 1944, 1998, 2014, 2015, 2016, 2017])
)

#year_2016_text
year_2016 = points.mark_text(align='left',baseline='middle',fontSize = 15,fontStyle = 'bold', dx=-40
                                                ).encode(text = "Year:N").transform_filter(
    alt.FieldOneOfPredicate(field='Year', oneOf = [2016])
)

#horisontal_line
line = alt.Chart(pd.DataFrame({'y': [0]})).mark_rule(color ="grey").encode(y='y')

#descriptive_text
des_text = alt.Chart({'values':[{'x': 2015, 'y': 0.1}]}).mark_text(
    text='HOTTER THAN THE\n1880-1899 AVERAGE\n\nCOLDER', lineBreak='\n',  fontSize=13, align='right').encode(x='x:Q', y='y:Q')

#title
title = alt.Chart({'values':[{'x': 1885, 'y': 0.95}]}).mark_text(align='left',
    text='Annual Global Surface Temperature,\nRelative To Late 19th Century Average', fontSize=15, lineBreak='\n',fontStyle ="bold").encode(x='x:Q', y='y:Q')


(points + years_text + line + year_2016 + des_text + title).properties(width = 900, height = 500, background = '#F9F9F9', padding = 25)


```
